**PEMROSESAN SEKUENSIAL ARRAY**

```python
arr = [10, 20, 30, 40, 50]
for elemen in arr:
    print(elemen)
```

```python
total = 0
for elemen in arr:
    total += elemen
print(total)  # Output: 150
```

```python
arr = [30,45,10,20,12,7,19,13,17,2,10,27,5,29]

nilai_maks = max(arr)
print("jadi nilai maksimun adalah:", nilai_maks)

nilai_min = min(arr)
print("jadi nilai minimun adalah:", nilai_min)

penjumlahan = sum(arr)
print("jadi total penjumlahan seluruh elemen:", penjumlahan)

jumlah_arr = len(arr)
print("jadi total jumlah seluruh elemen:", jumlah_arr)

rata_rata = penjumlahan/jumlah_arr
print("rata rata seluruh elemen:", rata_rata)
```

```python
# Soal 2 dengan bu Sevinurafni
T1 = [9, 12, 30, -1, 0, 4, -1, 3, 30, 14]
nilai_terbesar = max(T1)
print ("nilai terbesar dari T1 adalah", nilai_terbesar)

nilai_terkecil = min(T1)
print ("nilai terkecil dari T1 adalah", nilai_terkecil)

T2 = [1, 3, 5, 8, -12, 90, 3, 5]
nilai_terbesar = max(T2)
print ("nilai terbesar dari T2 adalah", nilai_terbesar)

nilai_terkecil = min(T2)
print ("nilai terkecil dari T2 adalah", nilai_terkecil)
```
